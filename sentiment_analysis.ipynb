```python
texts = texts[0:250]
labels = labels[0:250]
```

```python
len(labels)
```

```python
tokenizer = Tokenizer(num_words=1000, oov_token="<OOV>")
tokenizer.fit_on_texts(texts)
```

```python
sequesnce = tokenizer.texts_to_sequences(texts)
```

```python
sequesnce
```

```python
tokenizer.word_index
```

```python
sequesnce
```

```python
maxlen = 15
padded_sequence = pad_sequences(sequesnce, maxlen=15, padding='post')
```

```python
padded_sequence
```

```python
x_train, x_test, y_train, y_test = train_test_split(padded_sequence, labels, test_size=0.20, random_state=42)
```

```python
x_train
```

```python
# rnn_model = Sequential([
#     Embedding(input_dim=1000, output_dim=32),
#     SimpleRNN(128),
#     Dense(1, activation='sigmoid')
# ])
```

```python
rnn_model = Sequential([
    Embedding(input_dim=1000, output_dim=32),
    LSTM(128),
    Dense(1, activation='sigmoid')
])
```

```python
rnn_model.compile(optimizer='adam', loss = 'binary_crossentropy', metrics=['accuracy'])
print("Now I will start the training")
rnn_model.fit(np.array(x_train), np.array(y_train), epochs = 100, verbose=2)
```

```python
rnn_model.evaluate(np.array(x_test), np.array(y_test), verbose = 32)
```
